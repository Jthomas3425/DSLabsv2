### Step 1: Read the dataset and inspect its columns and 5-point statistics

### Step 4: Run a simple regression in `statsmodels` with TV as a predictor

```python
# import libraries

# build the formula 
f = 'sales~TV'
# create a fitted model in one line
model = smf.ols(formula=f, data=advert_data).fit()
```

### Step 5: Get regression diagnostics summary


```python
model.summary()
```

### Record your observations on "Goodness of fit"


Note here that the coefficients represent associations, not causations


### Step 6:  Draw a prediction line with data points omn a scatter plot for X (TV) and Y (Sales)

Hint: We can use `model.predict()` functions to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
tv_minmax = pd.DataFrame({'TV': [advert_data['TV'].min(), advert_data['TV'].max()]}) #Could pass in variables directly into the model declaration above
tv_minmax
# make predictions for those x values and store them
predictions = model.predict(tv_minmax)
# first, plot the observed data and the least squares line
advert_data.plot(kind='scatter', x='TV', y='sales')
plt.plot(tv_minmax, predictions, color='red')
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.regressionplots.plot_regress_exog(model, 'TV', fig=fig)
plt.show()
```

### Step 7: Visualize the error term for variance and heteroscedasticity

```python
# Record Your observations on residuals
```

```python
f_radio = 'sales~radio'
radio_model = ols(formula=f_radio, data=advert_data).fit()
radio_model.summary()

#Regression Line
radio_minmax = pd.DataFrame({'radio': [advert_data['radio'].min(), advert_data['radio'].max()]})
radio_pred = radio_model.predict(radio_minmax)
advert_data.plot(kind='scatter', x='radio', y='sales')
plt.plot(radio_minmax, radio_pred, color='red')
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.regressionplots.plot_regress_exog(radio_model, 'radio', fig=fig)
plt.show()
```
